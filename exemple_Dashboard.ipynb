```python
#on crée la fonction qui recherche ce qu'on veut
def annee(texte) :
    txt = BeautifulSoup(str(texte))
    
    #parcourir les balises
#     em_box = txt.find("div", {"class": "article-page"})
#     em_box = em_box.find("div", {"class": "article-details"})
#     em_box = em_box.find("header", {"class": "heading"})
#     em_box = em_box.find("div", {"class": "article-citation"})
    em_box = txt.select('div.article-citation')[0].text.strip()

    #chercher l'info interessante
    date = re.findall('[0-9]*[0-9][0-9] *[A-Z]*[a-z]* *[0-9]*;',str(em_box))
    return(date)
```

```python
#on teste la fonction sur quelques ligne du dataframe
dates = {}
for i in range(0,10):
    url = df.iloc[i,1]
    texte = df.iloc[i,0]
    
    date = annee(texte)
    dates[url] = date
```

```python
#on regarde si ça a marché, sinon on ajuste
dates
```

```python
#si ça marche on applique la fonction sur tout le dataframe
liste = [annee(df.iloc[i,0]) for i in range(0,len(df))]
```

```python
#on ajoute cette nouvelle colonne dans le dataframe
df['Annee'] = liste
```

```python
#on vérifie que ça semble avoir marché
df.head()
```

```python
#on enregistre le dataframe au format csv
df.to_csv('articles_v1.csv', index = False)
```

```python
#on crée la fonction qui recherche ce qu'on veut
def NomPrenom(texte) :
    txt = BeautifulSoup(str(texte))
    auteurs = []
    #parcourir les balises
    em_box = txt.find("div", {"class": "article-page"})
    em_box = em_box.find("div", {"class": "article-details"})
    em_box = em_box.find("header", {"class": "heading"})
    em_box = em_box.find("div", {"class": "authors-list"})
    if ((em_box) == None) :
        return('walou')
    em_box = em_box.find_all("a",href=True)
    #chercher l'info interessante
    i = 0
    while i < len(em_box) :
        auteurs.append(list(em_box[i])[0])
        i+=2
    return(auteurs)
```

```python
#on teste la fonction sur quelques ligne du dataframe
NP = {}
for i in range(0,10):
    url = df.iloc[i,1]
    texte = df.iloc[i,0]
    
    np = NomPrenom(texte)
    NP[url] = np
```
